```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# La fonction donnee par le prof
def f(x):
    return (x+2.5)**2
```

```python
# Nous construisons pour voir la courbe
plt.figure(figsize=(8,6))
x = np.linspace(-10,5,200)
plt.plot(x,f(x))

plt.xlabel('x')
plt.ylabel("f(x)")
plt.show()
```

```python
# Nous voyons que le minimum se trouve a peu pres entre -5 et 0
# Nous allons ecrire l'algorithm de Dichotomy pour le trouver
```

```python
def dichotomy(f,a,b,eps=0.05):
    while abs(b-a) > eps:
        c = (b+a)/2 # Le milieu de l'inteval [a,b]
        
        x1 = c - 0.001 # Un point a gauche du centre
        x2 = c + 0.001 # Un point a droite du centre
        
        if(f(x1) > f(x2)):
            a = x1 # On bouge la borne de gauche
        else:
            b = x2 # On bouche la borne de droite
    
    x_sol = (b+a)/2 # Des que l'interval [a,b] devient tres petit c'est-a-dire (b-a) < eps, on prend son milieu comme solution
    
    return x_sol
```

```python
print(dichotomy(f,a=-5,b=0))
```

```python
# Si on veut plus de precision on prend eps encore plus petit, par exemple eps = 0.01
print(dichotomy(f,a=-5,b=0,eps=0.01))
```

```python
# On peut construire le point sur le graphique
x_sol = dichotomy(f,a=-5,b=0,eps=0.01)
y_sol = f(x_sol)

plt.figure(figsize=(8,6))
x = np.linspace(-5,0,200)
plt.plot(x,f(x))
plt.scatter([x_sol],[y_sol],color='red',label='Решение задачи оптимизации')

plt.xlabel('x')
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
