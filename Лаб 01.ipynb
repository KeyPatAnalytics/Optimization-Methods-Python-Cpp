```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# La fonction donnee
def f(x):
    return (x+2.5)**2
```

```python
# Nous construisons le graphic
# Juste pour regarder la forme de la fonction
xx = np.linspace(-10,5,200)
plt.plot(xx,f(xx))
plt.show()
```

```python
# L'algorithm de swenn (Алгоритм свенна) permet de trouver
# L'interval sur lequel se trouve un minimum

def swenn(f,x0,t): # х0 - начальная точка, t - шаг
    x1 = x0
    x2 = x1 + t
    if f(x1) < f(x2): # мы меняем направление в сторону убывания
        t *= -1.0
        x2 = x1 + t
    x3 = x2 + t
    
    while f(x2) > f(x3): # Мы продолжаем пока функция не станет возрастать
        t = 2 * t
        x1, x2, x3 = x2, x3, x3 + t
    
    return (x1, x3) if x1 < x3 else (x3, x1) # Мы возрашаем интервал локализации минимума
```

```python
a, b = swenn(f,x0=6,t=1)
print(f"Интервал локализации минимума: a={a}, b={b}")
```

```python
xx = np.linspace(-10,5,200)
plt.plot(xx,f(xx),label = 'Целевая функция')
plt.plot([a,b],[0,0],'r-', label = "Интервал локализации минимума")

plt.legend()
plt.show()
```
